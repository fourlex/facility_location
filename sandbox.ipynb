```python
import pandas as pd
from facility_location import one_center_plane, geometric_median_plane, geometric_median_sphere
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
df = pd.read_csv('5ka_moscow.csv', names=('lat', 'lon', 'w'))
df
```

```python
p = df.values[:,:2]
w = df.values[:,2]
y_plane, _ = geometric_median_plane(p, w)
y_sphere, _ = geometric_median_sphere(p, w)
y_one_center, r = one_center_plane(p)
fig = plt.figure(figsize=(5,5), dpi=150)
plt.scatter(df['lon'], df['lat'], c='black', marker='.', label='facilities', s=1)
plt.scatter(y_plane[1], y_plane[0], c='r', marker='+', label='plane median')
plt.scatter(y_sphere[1], y_sphere[0], c='g', marker='+', label='sphere median')
plt.scatter(y_one_center[1], y_one_center[0], c='b', marker='+', label='plane 1-center')
plt.gca().add_patch(plt.Circle((y_one_center[1], y_one_center[0]), radius=r, alpha=0.05))
plt.xlabel('Lon')
plt.ylabel('Lat')
plt.legend()
```
